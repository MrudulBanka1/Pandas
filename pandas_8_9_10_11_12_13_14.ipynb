```python
#Merge Data Frames
```

```python
df1 =pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
})
df1
```

```python
df2 =pd.DataFrame({
    "city":["chicago","new york","orlando"],
    "humididty":[68,65,75]
})
df2
```

```python
df3 = pd.merge(df1,df2, on = "city")
df3
```

```python
df1 =pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature":[21,14,35,40],
})
df1
```

```python
df2 =pd.DataFrame({
    "city":["chicago","new york","San Fransisco"],
    "humididty":[68,65,71]
})
df2
```

```python
df3 = pd.merge(df1,df2, on = "city")
df3
```

```python
df3 = pd.merge(df1,df2, on = "city", how = "outer")
df3
```

```python
df3 = pd.merge(df1,df2, on = "city", how = "left")
df3
```

```python
df3 = pd.merge(df1,df2, on = "city", how = "right")
df3
```

```python
df3 = pd.merge(df1,df2, on = "city", how = "outer", indicator = True)
df3
```

```python
df1 =pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature":[21,14,35],
    "humididty":[68,65,75]
})
df1
```

```python
df2 =pd.DataFrame({
    "city":["Frisco","chicago","new york"],
    "temperature":[21,14,35],
    "humididty":[68,65,75]
})
df2
```

```python
df3 = pd.merge(df1,df2, on = "city", suffixes =('_left','_right'))
df3
```

```python
#CrossTab for statistiacl modelling
import pandas as pd
df = pd.read_excel("G:\\python_demo1\\pandas\\13_crosstab\\survey.xls")
df
```

```python
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Nationality, df.Handedness, margins = True)
```

```python
pd.crosstab(df.Sex,[df.Nationality, df.Handedness], margins = True)
```

```python
pd.crosstab([df.Sex,df.Nationality], df.Handedness, margins = True)
```

```python
pd.crosstab(df.Nationality, df.Handedness, normalize = "index")
```

```python
import numpy as np
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average )
```
